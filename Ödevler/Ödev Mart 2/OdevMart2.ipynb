# NumPy Kütüphanesi: 10 Fonksiyon Açıklaması ve Kullanım Örnekleri
Bu notebook, NumPy kütüphanesindeki 10 önemli fonksiyonu açıklayarak her birinin kullanımına dair örnekler sunmaktadır.

### Gerekli Kütüphaneler
Aşağıdaki kod hücresini çalıştırarak NumPy kütüphanesini içe aktarabilirsiniz.

```python
import numpy as np
```

## 1. np.array()
NumPy dizisi oluşturmak için kullanılır.

```python
# Tek boyutlu bir NumPy dizisi oluşturma
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

## 2. np.zeros()
Belirtilen boyutta tüm elemanları 0 olan bir dizi oluşturur.

```python
# 3x3 boyutunda sıfırlardan oluşan bir matris
zeros_array = np.zeros((3,3))
print(zeros_array)
```

## 3. np.ones()
Belirtilen boyutta tüm elemanları 1 olan bir dizi oluşturur.

```python
# 2x4 boyutunda birlerden oluşan bir matris
ones_array = np.ones((2,4))
print(ones_array)
```

## 4. np.arange()
Belirtilen başlangıç, bitiş ve adım değeri ile bir dizi oluşturur.

```python
# 0'dan 10'a kadar 2'şer artan dizi oluşturma
arange_array = np.arange(0, 10, 2)
print(arange_array)
```

## 5. np.linspace()
Belirtilen iki değer arasında eşit aralıklı sayılar oluşturur.

```python
# 0 ile 1 arasında 5 eşit bölünmüş sayı oluşturma
linspace_array = np.linspace(0, 1, 5)
print(linspace_array)
```

## 6. np.random.rand()
0 ile 1 arasında rastgele sayılardan oluşan bir dizi oluşturur.

```python
# 3x3 boyutunda rastgele sayılar içeren bir matris
random_array = np.random.rand(3,3)
print(random_array)
```

## 7. np.reshape()
Dizinin boyutlarını değiştirmek için kullanılır.

```python
# 1D diziyi 2D'ye dönüştürme
arr = np.arange(6)
reshaped_arr = arr.reshape(2,3)
print(reshaped_arr)
```

## 8. np.mean()
Bir dizinin ortalamasını hesaplar.

```python
# Ortalama hesaplama
arr = np.array([10, 20, 30, 40, 50])
mean_value = np.mean(arr)
print(mean_value)
```

## 9. np.std()
Bir dizinin standart sapmasını hesaplar.

```python
# Standart sapma hesaplama
std_value = np.std(arr)
print(std_value)
```

## 10. np.dot()
İki dizinin matris çarpımını gerçekleştirir.

```python
# İki matrisin çarpımı
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
dot_product = np.dot(A, B)
print(dot_product)
```
