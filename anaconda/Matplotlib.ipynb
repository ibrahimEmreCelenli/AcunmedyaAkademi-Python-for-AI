# Matplotlib (Veri Görselleştirme)

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y, marker="s", linestyle="dotted", color="b", label="Satışlar")
plt.legend() #Sol Üstteki bilgi kutusunu çıkarıyor
plt.xlabel("Günler")
plt.ylabel("Satışlar")
plt.grid()
plt.title("Günlük Satış Rakamları")
plt.show()
```
