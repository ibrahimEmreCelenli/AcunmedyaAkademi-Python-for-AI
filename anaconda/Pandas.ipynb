# Boş Veri Doldurma

```python
df3.isnull().sum()
```

```python
df3["Country"] = df3["Country"].fillna(df3["Country"].mode()[0])
df3.isnull().sum()
```

#### Yukarıda Country kısmındaki boş olan 8 veri en sık kullanılan Country verisi ile dolduruldu. Aşağıda ise farklı bir şekilde Boxes Shipped kısmı dolduruldu

```python
df3["Boxes Shipped"].fillna(df3["Boxes Shipped"].mean(), inplace=True)
df3.fillna({["Boxes Shipped"]:df3["Boxes Shipped"].mean})
df3.isnull().sum()
```

#### Amount kısmının tipi object (dolar ve virgül işaretleri var) bunu önce stringe çevirip "$" ve "," işaretlerini kaldırıp float'a çevirmemiz lazım

```python
df3["Amount"] = df3["Amount"].astype(str).str.replace("[$,]","",regex=True).astype(float)
```

```python
df3["Amount"] = df3["Amount"].fillna(df3["Amount"].mean())
df3.isnull().sum()
```

#### Date kısmında tarih olduğundan üstündeki (veya yoksa altındaki) veriye göre dolduruyoruz

```python
df3["Date"] = df3["Date"].ffill().bfill()
df3.isnull().sum()
```

#### Product kısmında ise boşlukları her bir satıcının en çok satılan ürünü ile doldurduk

```python
df3["Product"] = df3["Product"].fillna(df3.groupby("Sales Person")["Product"].transform(pd.Series.mode))
df3.isnull().sum()
```
